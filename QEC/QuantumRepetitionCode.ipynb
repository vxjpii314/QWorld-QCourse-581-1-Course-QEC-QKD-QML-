```python
import stac
"""
You need to first install the library 'stac' for this code to work on your computer/IDE.
Installation Instruction is here: https://github.com/abdullahkhalids/stac/tree/main

"""

#Encoding Circuit
enc_circ = stac.Circuit.simple(3)
enc_circ.append('CX', 0, 1)
enc_circ.append('CX', 0, 2)
enc_circ.draw()
```

```python
#Decoding Circuit
synd_circ = stac.Circuit.simple(5)
synd_circ.append('CX', 0, 3)
synd_circ.append('CX', 1, 3)
synd_circ.append('TICK')
synd_circ.append('CX', 1, 4)
synd_circ.append('CX', 2, 4)
synd_circ.append('TICK')
synd_circ.append('M', 3)
synd_circ.append('M', 4)
synd_circ.draw()
```

```python
# Encoding + error + syndrome measurement
circ = stac.Circuit.simple(5)

#Set message to |1>
# circ.append('X', 0)

#Encode the logical qubit by entangling qubit 0 with qubits 1 and 2
circ.append('CX', 0, 1)
circ.append('CX', 0, 2)

#Introduce a single-qubit bit-flip error on qubit 0
circ.append('X', 0)

#Measure the first syndrome using ancilla 3
circ.append('CX', 0, 3)
circ.append('CX', 1, 3)
circ.append('TICK')

#Measure the second syndrome using ancilla 4
circ.append('CX', 1, 4)
circ.append('CX', 2, 4)
circ.append('TICK')

#Read out the ancilla 
circ.append('M', 3)
circ.append('M', 4)

#Sample the measurement results
circ.sample()
```
